```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set()
```

```python
#Set up hypothetical values of variables

#Mean weight of population surveyed
mean_p = 155
#Standard deviation
sd_p = 5
#Mean weight of population that goes to the gym
mean_gym = 185 
#Standard devation
sd_gym = 5
#Percentage of population that goes to the gym
gymperc = 0.30
#Size of total population
totalpopsize = 40000
```

```python
#Create two subgroups to separate gym-goers from non-gym-goers
non_gymgoers = np.random.normal(mean_p, sd_p, int(totalpopsize * (1 - gymperc)))
gymgoers = np.random.normal(mean_gym, sd_gym, int(totalpopsize * gymperc))

#Combine both subgroups to create the population
population = np.append(non_gymgoers, gymgoers)
```

```python
#Set up figure for plotting 
plt.figure(figsize = (10, 10))

#Set up subplot for non-gymgoers
plt.subplot(3,1,1)
sns.histplot(non_gymgoers, kde = True).set(title = 'Non-gymgoers only')
plt.xlim([140,200])

#Set up subpplot for gymgoers
plt.subplot(3,1,2)
sns.histplot(gymgoers, kde = True).set(title = 'Gymgoers only')
plt.xlim([140,200])

#Set up subplot for the combine population
plt.subplot(3,1,3)
sns.histplot(population, kde = True).set(title = 'Full population')
plt.axvline(x = np.mean(population))
plt.xlim([140,200])
plt.show()
```

# What if we sampled from the entire population?

We'll sample randomly from all members of the popuation

```python
#Simulation parameters
numSamps = 5000
sampSize = 50

#Sampling distribution of the mean for the entire population
mean_dist = np.empty(numSamps)
for i in range(numSamps):
    random_people = np.random.choice(population, sampSize)
    mean_dist[i] = np.mean(random_people)
    
#Plot the population and the biased sampling distribution 
plt.figure(figsize = (10,8))

#Plot the population again
plt.subplot(2,1,1)
sns.histplot(population, kde = True).set(title = 'Full population')
plt.axvline(x = np.mean(population))
plt.xlim([140,200])

#Plot the sampling distribution of the mean
plt.subplot(2,1,2)
sns.histplot(mean_dist, kde = True).set(title = 'Sampling distribution of the mean weight of entire population')
plt.axvline(x = np.mean(population))
plt.axvline(x = np.mean(mean_dist), color = 'black')
plt.xlim([140,200])

plt.show()
```

# What if we take a non-representive sampling?
What happens if we only go to the gym for our weight samples, instead of the entire population?

```python
#Simulation parameters
numSamps = 5000
sampSize = 50

#Sampling distribution of the mean for the entire population
mean_dist = np.empty(numSamps)
for i in range(numSamps):
    random_people = np.random.choice(gymgoers, sampSize)
    mean_dist[i] = np.mean(random_people)

#Plot population once non-representative sampling distribution
plt.figure(figsize = (10,8))

#Plot the entire population once again
plt.subplot(2,1,1)
sns.histplot(population, kde = True).set(title = 'Total population')
plt.axvline(np.mean(population))
plt.xlim([140,200])

#Plot non-representative sampling distribution
plt.subplot(2,1,2)
sns.histplot(mean_dist, kde = True).set(title = 'Sampling distribution of the mean weight of gymgoers')
plt.axvline(np.mean(population))
plt.axvline(np.mean(mean_dist), color = 'Black')
plt.xlim([140,200])

plt.show()
```
